```python
import re
import csv
```

```python
csvfile = open('Seiyuu Nickname Bot Data.csv')
```

```python
csv_read = csv.DictReader(csvfile, delimiter=',')
```

```python
def nickname_stemmer(name_h, nickname_h):
    family_name = ""
    given_name = ""
    combinations = []
    try:
        family_name, given_name = name_h.split(" ")
    except ValueError:
        given_name = name_h
    print("Family_name: "+family_name)
    print("Given_name: "+given_name)
    print("Nickname: "+nickname_h)
    
    #permutations of family name, prioritise longer length permutations
    sub_family_name = [name.replace("~","") for name in sub_names(family_name)]
    sub_family_name.sort(key=len,reverse=True)
    for sub_name in sub_family_name:
        pos=""
        if nickname_h.startswith(sub_name) or nickname_h.endswith(sub_name):
            pos="family"
        if pos != "": 
            combinations.append([sub_name,pos])
            break
            
    #permutations of given name, prioritise longer length permutations
    sub_given_name = [name.replace("~","") for name in sub_names(given_name)]
    sub_given_name.sort(key=len,reverse=True)
    for sub_name in sub_given_name:
        pos=""
        if nickname_h.startswith(sub_name) or  nickname_h.endswith(sub_name):
            pos="given"
        if pos != "": 
            combinations.append([sub_name,pos])
            break
    return combinations
```

```python
def sub_names(name):
    """Generates substr permutations"""
    ngram = []
    for i in range(len(name)):
        ngram.append((name[0:i+1]+"~") if i != len(name)-1 else name[0:i+1])
        ngram.append(("~"+name[i:]) if i != 0 else name[i:])
    ngram = list(set(ngram))
    return ngram
```

```python
def nickname_parts(name_h, nickname_h):
    a = nickname_stemmer(name_h, nickname_h)
    #a[0] = name_stem, a[1] = pos
    if not a:
        return []
    else:
        stuff = []
        n = str(nickname_h)
        for combi in a:
            stem = str(combi[0]).replace("~","")
            t= re.sub(r'^'+stem+'|'+stem+'$', "~",n)
            stuff.append([combi[0],t,combi[1]])
        return stuff
```

```python
a = nickname_parts("さくら あやね","あやねる")
print(a)
```

```python
outf = open("outtt.csv", "a")
for row in csv_read:
    a = nickname_parts(str(row['name_h']),str(row['nickname_h']))
    if not a:
        outf.write(str(row['name'])+","+str(row['name_h'])+","+str(row['nickname_h'])+","+str(row['nickname'])+",none,,\n")
        print("Nickname has nothing to do with name\n")
    else:
        for b in a:
            print("from "+b[2]+" name: "+b[0]+" \nnickname appendage: "+b[1]+"\n")
            outf.write(str(row['name'])+","+str(row['name_h'])+","+str(row['nickname_h'])+","+str(row['nickname'])+","+b[2]+","+b[0]+","+b[1]+"\n")
```
